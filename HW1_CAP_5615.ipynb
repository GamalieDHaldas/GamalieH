#Question 1. Please use your own language to briefly explain the following concepts 

• **Machine Learning:** Think of machine learning as a computer science field where computers get the ability to learn patterns from data without us having to instruct them in detail. They use this learning to draw conclusions or make predictions. It's like we're enabling them to "learn from experience."

• **Features:** In the context of machine learning, features pertain to measurable attributes or properties of the phenomena being studied. For instance, if we are looking at a dataset about houses, the features could include square footage, number of bedrooms, age of the house, and so on. These features provide crucial input for machine learning algorithms.

•**Training data:** This is the information that we provide to a machine learning model during its learning phase. The training data includes both the input data and the corresponding correct output. The model tries to understand the relationship between inputs and outputs, with the goal of applying this understanding to new, unseen data.

• **Underfitting:** This happens when a machine learning model is too simple and doesn't capture the complexities or patterns in the data well enough. This leads to poor performance not only on the unseen data but also on the training data. It's like trying to fit a straight line to data that is better represented by a curve.

• **Overfitting:** In contrast to underfitting, overfitting is when the machine learning model pays too much attention to the noise or outliers in the training data. It becomes so complex that it performs really well on the training data, but doesn't generalize well to new data.

• **Confusion Matrix:** To see how well our model is doing, we can use a tool called a confusion matrix. It's basically a table that compares the actual and predicted classifications, showing us the model's hits and misses in terms of true positives, false positives, true negatives, and false negatives.

• **True Positive Rate and False Positive Rate:** When we're talking about a binary classification model, the True Positive Rate (or recall) tells us how well the model is at identifying positive cases correctly. On the other hand, the False Positive Rate tells us how often the model wrongly identifies a negative case as positive.

• **ROC (Receiver Operating Characteristic) Curve:** This is a graph we can use to see the performance of our binary classifier system. It plots the True Positive Rate against the False Positive Rate for different thresholds. Ideally, we want the curve to be close to the top left corner. The area under this curve, or AUC, gives us a single measure of the model's performance.

- **Explain how to use ROC curve to evaluate the performance of a classifier**

The ROC (Receiver Operating Characteristic) curve is a valuable tool for evaluating classifier performance. It plots the true positive rate (TPR) against the false positive rate (FPR) at different thresholds. A higher TPR and lower FPR indicate better performance. The area under the curve (AUC) summarizes overall performance, with a perfect classifier having an AUC of 1 and a random classifier having an AUC of 0.5. Analyzing the ROC curve and calculating the AUC allows for assessment and comparison of classifier effectiveness.

- **Why ROC curve is better than classification accuracy in assessing the classifier performance**

The ROC curve is more effective than classification accuracy in evaluating a classifier's competence. It provides a broad picture of the classifier's skill in discerning between positive and negative cases. 

Let's suppose you have a dataset with limited positive samples, say 99 negative and just one positive. A classifier might identify all as negative, which yields a 99% classification accuracy. This, however, does not truly reflect the classifier's ability as it fails to recognize the positive case. 

Therefore, accuracy isn't always an ideal performance measure. A more thorough understanding of the classifier's effectiveness can be drawn from the True Positive Rate (TPR) and False Positive Rate (FPR). In this situation, the TPR would be 0/1 (or 0) as it missed the positive sample, and the FPR would be 0/99 (or 0) since it correctly identified all the negatives. These figures would place the classifier at the origin of the ROC curve, indicating that it isn't up to the task.

Thus, the ROC curve is preferred when determining a classifier's proficiency, especially with skewed datasets. It doesn't just reflect the accuracy, but also evaluates the classifier's ability to accurately categorize both positive and negative instances.

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import nltk

nltk.download('punkt')

# Function to preprocess the text
def preprocess(text):
    text = text.lower()
    tokens = nltk.word_tokenize(text)
    tokens = [word for word in tokens if word.isalpha()]  # Remove punctuation marks
    return ' '.join(tokens)

# Read the CSV file
reviews = pd.read_csv("review100.csv")

# Preprocess the reviews
reviews['movie review'] = reviews['movie review'].apply(preprocess)

# Separate positive and negative reviews
positive_reviews = reviews[reviews['sentiment'] == 'positive']
negative_reviews = reviews[reviews['sentiment'] == 'negative']

# Calculate average length of reviews
positive_review_length = positive_reviews['movie review'].apply(lambda x: len(x.split())).mean()
negative_review_length = negative_reviews['movie review'].apply(lambda x: len(x.split())).mean()

print("Average length of positive reviews:", positive_review_length)
print("Average length of negative reviews:", negative_review_length)

# Get words from positive and negative reviews
positive_words = ' '.join(positive_reviews['movie review']).split()
negative_words = ' '.join(negative_reviews['movie review']).split()

# Count word frequencies for positive and negative reviews
positive_word_counts = Counter(positive_words)
negative_word_counts = Counter(negative_words)

# Remove specific punctuation marks
punctuation_marks = [',', '.']
positive_word_counts = {word: count for word, count in positive_word_counts.items() if word not in punctuation_marks}
negative_word_counts = {word: count for word, count in negative_word_counts.items() if word not in punctuation_marks}

# Get top 20 words for positive and negative reviews
top_positive_words = Counter(positive_word_counts).most_common(20)
top_negative_words = Counter(negative_word_counts).most_common(20)

# Print top words for positive reviews
print("\nTop 20 most frequent words in positive reviews:")
for word, frequency in top_positive_words:
    print(word, "-", frequency)

# Print top words for negative reviews
print("\nTop 20 most frequent words in negative reviews:")
for word, frequency in top_negative_words:
    print(word, "-", frequency)

# Prepare data for plotting
words_positive = [word[0] for word in top_positive_words]
counts_positive = [word[1] for word in top_positive_words]

words_negative = [word[0] for word in top_negative_words]
counts_negative = [word[1] for word in top_negative_words]

# Create DataFrames for positive and negative words
df_positive = pd.DataFrame(list(zip(words_positive, counts_positive)), columns=['Words', 'Counts'])
df_negative = pd.DataFrame(list(zip(words_negative, counts_negative)), columns=['Words', 'Counts'])

# Merge positive and negative DataFrames
df = pd.merge(df_positive, df_negative, on='Words', how='outer').fillna(0)
df = df.set_index('Words')
df.columns = ['Positive', 'Negative']

# Plot the stacked bar chart
df.plot(kind='bar', stacked=True, figsize=(10, 7))
plt.title('Top 20 Words in Positive and Negative Reviews')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```
